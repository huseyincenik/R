```r
# H0 : degisken normaldir.
# H1 : degisken normal degildir.

# p > alfa ise H0 red, p < alfa ise H0 kabul ediliyor

check_normality <- function(data, alpha = 0.05){
    # Apply the Shapiro-Wilk Test
    shapiro_result <- shapiro.test(data)
    
    # Print the p-value
    cat("Shapiro-Wilk Test p-value: ", shapiro_result$p.value, "\n")
    
    # Evaluate the test result with an if condition
    if (shapiro_result$p.value < alpha){
        cat("Reject the null hypothesis. The data is not normally distributed. \n")
    }
    else {
        cat("Fail to reject the null hypothesis. The data is normally distributed. \n")
    }
    
}
```

```r
check_normality(subset(new_df, group = G1)$RW);check_normality(subset(new_df, group = G2)$RW);
check_normality(subset(new_df, group = G3)$RW);check_normality(subset(new_df, group = G4)$RW)
# shapiro.test(subset(new_df, group = G1)$RW)
```

```r
# Grupları belirle
groups <- unique(new_df$group)

# Grafik boyutunu ayarla
# mar: c(sol, üst, sağ, alt) boşlukları belirtir
par(mfrow = c(ceiling(length(groups)/2), 2), mar = c(5, 4, 4, 2) + 0.1)

# Her bir grup için qqplot'u çiz
for (i in 1:length(groups)) {
  group_data <- subset(new_df, group == groups[i])$RW
  sirali <- sort(group_data)
  n <- length(group_data)
  j <- 1:n
  probability <- (j - 1/2) / n
  q_p <- qnorm(p = probability)
  
  # Grafik çizimi
  qqplot(x = q_p, y = sirali, main = paste("QQ Plot for Group", groups[i]), xlab = "Theoretical Quantiles", ylab = "Sample Quantiles")
}
```

```r
n_G1 = nrow(G1)
n_G2 = nrow(G2)
n_G3 = nrow(G3)
n_G4 = nrow(G4)
```

```r
# Anova -- RearWidth.
SSB = (mean_G1[2] - mean_general[2])^2*n_G1 +
      (mean_G2[2] - mean_general[2])^2*n_G2 +
      (mean_G3[2] - mean_general[2])^2*n_G3 +
      (mean_G4[2] - mean_general[2])^2*n_G4

# Gruplar arasi kareler toplami (GAKT)
# Her bir olcumun genel ortalamdan farklarinin kareler toplamidir.
SSB
```

```r
SSW = sum((G1[,2] - mean_G1[2])^2) +
      sum((G2[,2] - mean_G2[2])^2) +
      sum((G3[,2] - mean_G3[2])^2) + 
      sum((G4[,2] - mean_G4[2])^2) 
# Grup ici kareler toplami (GIKT)
# Her grup ortalamasinin genel ortalamadan farklarinin karelerinin toplamidir.
SSW
```

```r
# k grup sayisi, df_B Gruplar arasi serbestlik derecesi, df_W Gruplar ici serbestlik derecesi olmak uzere;
k = 4; df_B = k - 1;df_W = nrow(new_df) - k
```
